```python
import numpy as np 
import pandas as pd
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt

import json

from pathlib import Path
from shapely.geometry import Point
from shapely.geometry import shape

from functions import ( fill_hole, winds_split, timed_weather, get_geometry_APPA, SET_time, classify_pollutant, add_eaqi,
                        get_minute, get_hour, get_day, get_month, get_year)
```

```python
data_path = Path('../data/raw')
APPA_path = Path('../data/external')

files = {'grid':'trentino-grid.geojson',
         'adm_reg':'administrative_regions_Trentino.json',
         'weather':'meteotrentino-weather-station-data.json',
         'precip':'precipitation-trentino.csv',
         'precip-avail':'precipitation-trentino-data-availability.csv',
         'SET-1':'SET-nov-2013.csv',
         'SET-2':'SET-dec-2013.csv',
         'SET-lines':'line.csv',
         'twitter':'social-pulse-trentino.geojson'}

APPA_files = {'data_pol':'APPA_extended.csv',
              'pos':'posizioni_stazioni_appa.csv'}

grid_df = gpd.read_file(data_path / files['grid'])
```

### MeteoTrentino, ARPA

```python
with open(data_path / files['weather']) as f:
    weather_json = json.load(f)

weather = gpd.GeoDataFrame(weather_json['features'])

weather = fill_hole(weather)

weather_df = winds_split(weather)

weather_df['geometry'] = weather_df['geomPoint.geom'].apply(lambda x:Point(x['coordinates'][0], x['coordinates'][1]))
weather_df.drop(columns=['geomPoint.geom'],inplace=True)

t_weather_df = timed_weather(weather_df)

del weather
del weather_df
```

```python
t_weather_df
```

### APPA

```python
APPA_df = pd.read_csv(APPA_path / APPA_files['data_pol'], encoding='latin1')

APPA_df['Valore'] = pd.to_numeric(APPA_df['Valore'], errors='coerce')
APPA_df['Ora'] = pd.to_numeric(APPA_df['Ora'], errors='coerce')
```

```python
APPA_pos = pd.read_csv(APPA_path / APPA_files['pos'])
APPA_pos[['lat', 'lon']] = APPA_pos['Posizione'].str.split(',', expand=True)
APPA_pos = APPA_pos.drop('Posizione', axis=1)

data_names = ['Piana Rotaliana', 'Riva del Garda', 'Monte Gaza', 'Parco S. Chiara', 'Rovereto', 'Via Bolzano','A22 (Avio)', 'Borgo Valsugana']

APPA_pos['data names'] = data_names
```

```python
APPA_df['geometry'] = APPA_df['Stazione'].map(get_geometry_APPA(APPA_df, APPA_pos))
del APPA_pos

APPA_df['Inquinante'] = APPA_df['Inquinante'].replace({'Biossido Zolfo': 'SO2', 'Biossido di Azoto': 'NO2', 'Ossido di Carbonio': 'CO', 'Ozono': 'O3'})
```

```python
APPA_df
```

```python
# dobbiamo riordinare il dataframe in modo da adattarci al formato "cronologico" usato per weather
APPA_df['station_pollutant'] = (APPA_df['Stazione'] + '_' + APPA_df['Inquinante'])

APPA_df = APPA_df.pivot_table(index=['Data', 'Ora'], columns='station_pollutant', values='Valore').reset_index()

APPA_df
```

```python
# aggiungiamo una colonna per l'indice di qualità dell'aria
display(Image(filename='../images/TABELLA_APPA.png'))
```

```python
APPA_df = add_eaqi(APPA_df)
APPA_df
```

### SET

```python
SET_1_df         = pd.read_csv(data_path / files['SET-1'], header=None)
SET_2_df         = pd.read_csv(data_path / files['SET-2'], header=None)
SET_l_df         = pd.read_csv(data_path / files['SET-lines'])

SET_1_df.rename(columns={0: 'line_id', 1: 'time', 2: 'value'}, inplace=True)
SET_2_df.rename(columns={0: 'line_id', 1: 'time', 2: 'value'}, inplace=True)
```

```python
SET_l_df
```

```python
# prima di tutto estraiamo il numero di utenze della rete elettrica per ciscuna linea e casella della griglia
users = SET_l_df.groupby("LINESET")["NR_UBICAZIONI"].sum().reset_index()                                              \
    .rename(columns={"NR_UBICAZIONI": "UBICAZIONI_PER_LINEA"})  \
    .set_index("LINESET")
```

```python
# visto che sono feature utili, separiamo il tempo in giorno, orario e giorno della settimana, poi uniamo i due dataframe

# day va da 0 a 60
#weekday va da 0 (lunedi') a 6 (domenica)

# QUESTA CELLA E' DA SPOSTARE

SET_1_df = SET_time(SET_1_df)
SET_2_df = SET_time(SET_2_df)

SET_2_df['day'] += 30

SET_df = pd.concat([SET_1_df, SET_2_df], axis=0, ignore_index=True)

del SET_1_df
del SET_2_df

SET_df = SET_df[['line_id', 'day', 'time', 'weekday', 'value']]
```
