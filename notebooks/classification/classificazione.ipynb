# Classificazione
Dopo aver preparato il DataFrame nel notebook 'EDA.ipynb', vogliamo ora provare a predire l'indice di qualità dell'aria ad una certa ora di un certo giorno, usando come riferimento i dati precedenti a questa ora e giorno.
L'indice di qualità dell'aria (EAQI) può assumere quattro valori diversi: 'Buono', 'Discreto', 'Moderato', 'Scadente', 'Pessimo', nel nostro DataFrame indicati rispettivamente come 'good', 'fair', 'moderate', 'poor', 'awful'. Il problema è dunque di carattere multiclasse.

La classificazione fa parte della più grande branca del ML 'Supervised Learning'. In classe abbiamo visto diversi algoritmi per la classificazione. Quelli utili nel caso della multiclass classification sono il <u>Logistic Regression con soft max</u> ed il <u>Random Forest</u>
